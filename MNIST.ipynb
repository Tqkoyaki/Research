```python
import torch
import torchvision
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
```

```python
mnist = torchvision.datasets.MNIST(
    root = 'data/',
    download = True,
    train = True,
    transform = torchvision.transforms.ToTensor()
)
```

```python
class ImageClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.f = nn.Sequential(
            nn.Conv2d(1, 32, 3, 1),
            nn.ReLU(),
            nn.Conv2d(32, 64, 3, 1),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.25),
            nn.Flatten(1),
            nn.Linear(9216, 128),
            nn.ReLU(),
            nn.Dropout2d(0.5),
            nn.Linear(128, 10)
        )
    
    def forward(self, x, y):
        return -nn.functional.cross_entropy(self.f(x), y, reduction='none')

    def classify(self, x):
        q = self.f(torch.as_tensor(x)[None, ...])[0]
        return nn.functional.softmax(q, dim=0)
```

```python
model = ImageClassifier()

EPOCHS = 5
BATCH_NUM = 1000

mnist_batched = torch.utils.data.DataLoader(mnist, batch_size=BATCH_NUM)
model.train(mode=True)
optimizer = optim.Adam(model.parameters())

for epoch in tqdm(range(EPOCHS)):
    for batch_num, (imgs, lbls) in enumerate(mnist_batched):
        optimizer.zero_grad()
        
        loglik = model(imgs, lbls)
        e = -torch.mean(loglik)
        
        e.backward()
        optimizer.step()


model.train(mode=False)
```

```python
img, _ = mnist[110]
model.classify(img)
```
